```python
fig, ax = plt.subplots(2, 4, figsize=(20, 10))

# Tampilkan plot box pada axis pertama
df_transactions.boxplot(column='TransactionID', ax=ax[0][0])

# Tampilkan plot box pada axis kedua
df_transactions.boxplot(column='CustomerID', ax=ax[0][1])

# Tampilkan plot box pada axis ketiga
df_transactions.boxplot(column='CardID', ax=ax[0][2])

# Tampilkan plot box pada axis keempat
df_transactions.boxplot(column='GasStationID', ax=ax[0][3])

# Tampilkan plot box pada axis kelima
df_transactions.boxplot(column='ProductID', ax=ax[1][0])

# Tampilkan plot box pada axis keenam
df_transactions.boxplot(column='Amount', ax=ax[1][1])

# Tampilkan plot box pada axis ketujuh
df_transactions.boxplot(column='Price', ax=ax[1][2])

plt.show()
```

```python
# mencari nilai mean
mean_CardID = df_transactions['CardID'].mean()
mean_ProductID = df_transactions['ProductID'].mean()
mean_Amount = df_transactions['Amount'].mean()
mean_Price = df_transactions['Price'].mean()

# mencari nilai standard deviation
std_CardID = df_transactions['CardID'].std()
std_ProductID = df_transactions['ProductID'].std()
std_Amount = df_transactions['Amount'].std()
std_Price = df_transactions['Price'].std()

# membuat kondisi untuk mengidentifikasi nilai outlier
condition_CardID = (df_transactions['CardID'] < (mean_CardID - 2*std_CardID)) | (df_transactions['CardID'] > (mean_CardID + 2*std_CardID))
condition_ProductID = (df_transactions['ProductID'] < (mean_ProductID - 2*std_ProductID)) | (df_transactions['ProductID'] > (mean_ProductID + 2*std_ProductID))
condition_Amount = (df_transactions['Amount'] < (mean_Amount - 2*std_Amount)) | (df_transactions['Amount'] > (mean_Amount + 2*std_Amount))
condition_Price = (df_transactions['Price'] < (mean_Price - 2*std_Price)) | (df_transactions['Price'] > (mean_Price + 2*std_Price))

# mengubah nilai outlier menjadi NaN
df_transactions['CardID'] = df_transactions['CardID'].where(~condition_CardID, np.nan)
df_transactions['ProductID'] = df_transactions['ProductID'].where(~condition_ProductID, np.nan)
df_transactions['Amount'] = df_transactions['Amount'].where(~condition_Amount, np.nan)
df_transactions['Price'] = df_transactions['Price'].where(~condition_Price, np.nan)
```

```python
# mencari median
median_CardID = df_transactions['CardID'].median()
median_ProductID = df_transactions['ProductID'].median()
median_amount = df_transactions['Amount'].median()
median_Price = df_transactions['Price'].median()

# mengisi nilai NaN dengan median
df_transactions['CardID'] = df_transactions['CardID'].fillna(median_CardID)
df_transactions['ProductID'] = df_transactions['ProductID'].fillna(median_ProductID)
df_transactions['Amount'] = df_transactions['Amount'].fillna(median_amount)
df_transactions['Price'] = df_transactions['Price'].fillna(median_Price)
```

```python
fig, ax = plt.subplots(2, 4, figsize=(20, 10))

# Tampilkan plot box pada axis pertama
df_transactions.boxplot(column='TransactionID', ax=ax[0][0])

# Tampilkan plot box pada axis kedua
df_transactions.boxplot(column='CustomerID', ax=ax[0][1])

# Tampilkan plot box pada axis ketiga
df_transactions.boxplot(column='CardID', ax=ax[0][2])

# Tampilkan plot box pada axis keempat
df_transactions.boxplot(column='GasStationID', ax=ax[0][3])

# Tampilkan plot box pada axis kelima
df_transactions.boxplot(column='ProductID', ax=ax[1][0])

# Tampilkan plot box pada axis keenam
df_transactions.boxplot(column='Amount', ax=ax[1][1])

# Tampilkan plot box pada axis ketujuh
df_transactions.boxplot(column='Price', ax=ax[1][2])

plt.show()
```

```python
# mencari nilai mean
mean_Consumption = df_yearmonth['Consumption'].mean()

# mencari nilai standard deviation
std_Consumption = df_yearmonth['Consumption'].std()

# membuat kondisi untuk mengidentifikasi nilai outlier
condition_Consumption = (df_yearmonth['Consumption'] < (mean_Consumption - 3*std_Consumption)) | (df_yearmonth['Consumption'] > (mean_Consumption + 2*std_Consumption))

# mengubah nilai outlier menjadi NaN
df_yearmonth['Consumption'] = df_yearmonth['Consumption'].where(~condition_Consumption, np.nan)
```

```python
# mencari median dari
median_Consumption = df_yearmonth['Consumption'].median()

# mengisi nilai NaN dengan median
df_yearmonth['Consumption'] = df_yearmonth['Consumption'].fillna(median_Consumption)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))

df_yearmonth.boxplot(column='CustomerID', ax=ax[0])
df_yearmonth.boxplot(column='Date', ax=ax[1])
df_yearmonth.boxplot(column='Consumption', ax=ax[2])

ax[0].set_title('Boxplot of CustomerID')
ax[1].set_title('Boxplot of Date')
ax[2].set_title('Boxplot of Consumption')

plt.show()
```

Visualisasi Pembelian Produk Setiap Jam

```python
dfc1 = pd.merge(left=df_transactions, right=df_products, left_on='ProductID', right_on='ProductID')
dfc1
```

```python
dfc1["Time"] = pd.to_datetime(dfc1["Time"])
dfc1["Hour"] = dfc1["Time"].dt.hour
```

```python
grouped_data = dfc1.groupby(["Hour", "Description"]).size().reset_index(name="counts")
grouped_data
```

```python
import seaborn as sns

sns.set_style("darkgrid")

fig = plt.figure(figsize = (25, 7))

for description in grouped_data["Description"].unique():
    d = grouped_data[grouped_data["Description"]==description]
    sns.lineplot(x=d["Hour"], y=d["counts"], label=description)

plt.xlabel("Jam")
plt.ylabel("Jumlah")
plt.title("Jumlah Transaksi Setiap Jam")
plt.legend()

plt.show()
```

Visualisasi Transaksi Berdasarkan Tanggal

```python
plt.figure(figsize=(10, 5))

grouped_data = dfc1.groupby(['Date', 'Description']).size().reset_index(name='counts')

# create pivot table for heatmap
pivot_table = grouped_data.pivot(index='Description', columns='Date', values='counts')

# create heatmap
sns.heatmap(pivot_table, annot=True, fmt=".2f")
plt.xlabel("Date")
plt.ylabel("Product")
plt.title("Transaksi Berdasarkan Tanggal")
plt.show()
```
